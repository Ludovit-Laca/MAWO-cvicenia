```
link = "https://www.ebay.com/sch/i.html?_nkw=iphone+12+pro+max&_sop=12"
stranka = requests.get(link)
```

```
tree = html.fromstring(stranka.content)
```

```
vysledok = tree.xpath("//div/span[@class='s-item__price']/text()")
vysledok
```

```
ceny_nove = []
for c in vysledok:
  ceny_nove.append(float(c[1:].replace(',','')))

ceny_nove
```

```
import pandas
df = pandas.DataFrame(ceny_nove)
df.head()
```

```
df.describe()
```
