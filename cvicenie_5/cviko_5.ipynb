```
import requests
```

```
link = "https://www.sacred-texts.com/cla/aesop/index.htm"
f = requests.get(link)
```

```
bajky = f.text
```

```
from lxml import html
```

```
tree = html.fromstring(bajky)
```

```
vsetky_nazvy = tree.xpath("//a/text()")
vsetky_nazvy_spolu = ''

for nazov in vsetky_nazvy:
  vsetky_nazvy_spolu += ' ' + nazov.lower()
```

```
vsetky_nazvy_spolu
```

```
tokeny_bajky = word_tokenize(vsetky_nazvy_spolu)
```

```
frekvencia_bajky = FreqDist(tokeny_bajky)
```

```
frekvencia_bajky.most_common(20)
```

```
frekvencia_bajky['horse']
```

```
frekvencia_bajky['lion']
```

```
frekvencia_bajky['fish']
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```
sw = set (stopwords.words('english'))
sw.update(',')
```

```
nove_tokeny = []
for word in tokeny_bajky:
  if word not in sw: nove_tokeny.append(word)
```

```
frekvencia_bajky = FreqDist(nove_tokeny)
frekvencia_bajky.most_common(20)
```

```
%matplotlib inline
frekvencia_bajky.plot(30, cumulative=False)
```
