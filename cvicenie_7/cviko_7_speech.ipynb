```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import requests
```

```python
subor = requests.get('https://jozefkapusta.teacher.sk/web_content_mining/speeches/fico.txt',verify=False)
fico = subor.text
subor = requests.get('https://jozefkapusta.teacher.sk/web_content_mining/speeches/obama.txt',verify=False)
obama = subor.text
subor = requests.get('https://jozefkapusta.teacher.sk/web_content_mining/speeches/obama_bombings.txt',verify=False)
obama_bomb = subor.text
subor = requests.get('https://jozefkapusta.teacher.sk/web_content_mining/speeches/obama_hurricane.txt',verify=False)
obama_hur = subor.text
subor = requests.get('https://jozefkapusta.teacher.sk/web_content_mining/speeches/putin.txt',verify=False)
putin = subor.text
```

```python
prejavy = [fico, putin, obama, obama_bomb, obama_hur]
```

```python
vectorizer_count = CountVectorizer(stop_words='english')
model_pocetnosti = vectorizer_count.fit_transform(prejavy)
```

```python
import pandas
hlavicka = vectorizer_count.get_feature_names()
vypis_modelu = pandas.DataFrame(model_pocetnosti.toarray(),
                index=['fico','putin','obama','obama_bomb','obama_hur'],
                columns=hlavicka)
vypis_modelu
```

```python
vectorizer_tfidf = TfidfVectorizer(stop_words='english')
model_tfidf = vectorizer_tfidf.fit_transform(prejavy)
```

```python
import pandas
hlavicka = vectorizer_tfidf.get_feature_names()
vypis_modelu = pandas.DataFrame(model_tfidf.toarray(),
                index=['fico','putin','obama','obama_bomb','obama_hur'],
                columns=hlavicka)
vypis_modelu
```
