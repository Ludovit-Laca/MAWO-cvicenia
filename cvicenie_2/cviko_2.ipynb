```
prvy_text = "In the beginning was the Word , and the Word was with God , and the Word was God ."
```

```
bible = prvy_text.split(" ")
bible
```

```
len(bible)
```

```
# set will not have duplicates
mnozina = set(bible)
mnozina
```

```
len(mnozina)
```

```
# all lower letters
[w for w in bible if w.islower()]
```

```
#all upper letters
[w for w in bible if w.isupper()]
```

```
# words with capital letters
[w for w in bible if w.istitle()]
```

```
# alphabetic
[w for w in bible if w.isalpha()]
```

```
# digits
[w for w in bible if w.isdigit()]
```

```
# alpha numeric
[w for w in bible if w.isalnum()]
```
