```python
positive_file = open("/content/drive/MyDrive/Škola/MAWO/cvicenie_6/positive_word.txt").read()
negative_file = open("/content/drive/MyDrive/Škola/MAWO/cvicenie_6/negative_word.txt").read()
```

```python
positive_words = positive_file.splitlines()
negative_words = negative_file.splitlines()
```

```python
veta = 'Sushi was great, the food was awesome, but the service was terrible.'
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokeny = word_tokenize(veta)
```

```python
tokeny
```

```python
n_positive = 0
n_negative = 0
for token in tokeny:
    if token in positive_words:
        n_positive +=1
    if token in negative_words:
        n_negative += 1

print(f"Number of negative words {n_negative}\nNumber of positive words {n_positive}")
if n_positive > n_negative:
    print("Sentense has positive meaning")
else:
    print("Sentense has negative meaning")
```

```python
fiona = "hrdinstvo hrdinstvo hrdinstvo hrdinstvo závisť závisť krása krása"
shrek = "hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo hrdinstvo závisť závisť"
krason = "hrdinstvo závisť závisť krása krása krása krása krása krása"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
korpus_count = vectorizer.fit_transform([fiona,shrek,krason])
```

```python
import pandas
```

```python
print(pandas.DataFrame(korpus_count.A,columns = vectorizer.get_feature_names(), index = ['fiona', 'shrek', 'krason']))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
korpus_tfidf= vectorizer.fit_transform([fiona,shrek,krason])
```

```python
print(pandas.DataFrame(korpus_tfidf.A,columns = vectorizer.get_feature_names(), index = ['fiona', 'shrek', 'krason']))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
```

```python
euclidean_distances(korpus_count[0:1], korpus_count)
```

```python
euclidean_distances(korpus_tfidf[0:1], korpus_tfidf)
```

```python
cosine_similarity(korpus_tfidf[0:1], korpus_tfidf)
```

```python
cosine_similarity(korpus_count[0:1], korpus_count)
```
