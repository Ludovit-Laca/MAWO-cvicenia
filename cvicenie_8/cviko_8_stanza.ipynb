```python
import stanza
import requests, json
```

```python
stanza.download('sk')
```

```python
nlp = stanza.Pipeline(lang='sk')
```

```python
veta = "Spadla hruška zelená, rozbila si kolená. Teraz plače na zemi, dobré lieky dajte mi."
```

```python
dokument = nlp(veta)
```

```python
dokument
```

```python
po_lematizacii = ''
for vety in dokument.sentences:
  for slovo in vety.words:
    po_lematizacii += slovo.lemma + ' '
```

```python
po_lematizacii
```

```python
po_lematizacii = ''
for vety in dokument.sentences:
  for slovo in vety.words:
    print(slovo.text + ' - ' + slovo.upos + ' - ' + slovo.xpos)
```
