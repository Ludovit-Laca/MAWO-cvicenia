```python
from sklearn.model_selection import train_test_split
```

```python
y = spravy["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(spravy["text"], y, test_size = 0.2, random_state = 33)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words="english", ngram_range = (1,3))
```

```python
X_train_tfid = tfidf_vectorizer.fit_transform(X_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB().fit(X_train_tfid, y_train)
```

```python
X_test_tfid = tfidf_vectorizer.transform(X_test)
```

```python
X_test_tfid
```

```python
y_pred = clf.predict(X_test_tfid)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
print("Presnosť: ", metrics.accuracy_score(y_test, y_pred))
```
